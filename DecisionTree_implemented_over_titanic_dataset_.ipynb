```python
import pandas as pd
content=pd.read_csv("/ml.csv")
```

```python
content.describe()
```

```python
content.head()
```

```python
content['Name']
```

```python
inputs=content.drop('Survived',axis='columns')
target=content['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
p_class = LabelEncoder()
s_sex = LabelEncoder()
a_age = LabelEncoder()
f_fare = LabelEncoder()
```

```python
inputs
```

```python
inputs['p_class']=p_class.fit_transform(inputs['Pclass'])
inputs['s_sex']=s_sex.fit_transform(inputs['Sex'])
inputs['a_age']=a_age.fit_transform(inputs['Age'])
inputs['f_fare']=f_fare.fit_transform(inputs['Fare'])
```

```python
inputs
```

```python
inputs_n=inputs.drop(['Pclass','Age','Sex','Fare','Ticket','Cabin','Embarked','Parch','SibSp'],axis='columns')
```

```python
inputs_n
```

```python
inputs_n=inputs_n.drop(['Name','PassengerId'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,1,28,18]])
```
